```python
import numpy as np
import pandas as pd
import os
```

```python
DATA_DIR='C:/Users/gyang/Documents/DS/data'
FILE_NAME='credit_card_default_TRAIN.csv'
data_path=os.path.join(DATA_DIR,FILE_NAME)
#print(data_path)
ccd=pd.read_csv(data_path,index_col='ID')
ccd.head()
```

```python
ccd.rename(columns=lambda x:x.lower(),inplace=True)
```

```python
ccd.head()
```

```python
## Target col Y :"default payment next month"
correlation_matrix = ccd.corr()
print(correlation_matrix)
```

```python
df1=correlation_matrix.stack().reset_index()
```

```python
df1
```

```python
# pick out corr with Y
depvar=df1[df1['level_1']=='default payment next month']
print(depvar)
```

```python
df1
```

```python
df_inter=df1.merge(depvar,left_on='level_0', right_on='level_0', how='left')
```

```python
df=df_inter.merge(depvar,left_on='level_1_x', right_on='level_0', how='left').drop(columns=['level_1', 'level_0_y'])
```

```python
df
```

```python
df.columns=["var1","var2","corr_var1_var2","depvar","corr_var1_Y","corr_var2_Y"]
df
```

```python
df.loc[abs(df["corr_var1_Y"])>abs(df["corr_var2_Y"]), 'badvar'] = df['var2']
df.loc[abs(df["corr_var1_Y"])<abs(df["corr_var2_Y"]), 'badvar'] = df['var1']
```
